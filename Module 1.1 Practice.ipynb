```python
import matplotlib.pyplot as plt
import math as m
from collections import Counter

data = [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]

jumlah_data = sum(data)
mean_value = jumlah_data/banyak_data
sorted_data = sorted(data)
banyak_data = len(sorted_data)
median_value = sorted_data[m.floor(banyak_data/2)]
print(f"mean : {mean_value}")
print(f"median : {median_value}\n")


counter = Counter(data)
mode_data = dict(counter)
max_counter = max(list(mode_data.values()))

mode_value = []
for k, v in mode_data.items() :
    print(f"{k} muncul {v} kali")
    if v == max_counter :
        mode_value.append(k)
        
print(f"{mode_value}\n")
```

```python
plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='blue')
plt.axvline(x=mean_value, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean_value:.2f}')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Mean')

plt.legend()
plt.show()
```

```python
plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='green')
plt.axvline(x=median_value, color='red', linestyle='dashed', linewidth=1, label=f'Median = {median_value:.2f}')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Median')

plt.legend()
plt.show()
```

```python
plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='green')

#plt.axvline(x=mode_value[0], color='red', linestyle='dashed', linewidth=1, label=f'Modus 1 = {mode_value[0]:.2f}')

for i in mode_value :
    plt.axvline(x=i, color='red', linestyle='dashed', linewidth=1, label=f'modus = {i:.2f}')


plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Modes')

plt.legend()
plt.show()
```

```python
data_uas = [95, 35, 50, 65, 70, 85, 25, 100, 80, 90, 45, 40, 35, 20, 75, 50, 40, 60, 80, 75]

jumlah_data_uas = sum(data_uas)
mean_value_uas = jumlah_data_uas/banyak_data_uas
sorted_data_uas = sorted(data_uas)
banyak_data_uas = len(sorted_data_uas)
index_tengah_uas = m.floor(banyak_data_uas/2)
median_uas1 = sorted_data_uas[index_tengah_uas]
median_uas2 = sorted_data_uas[index_tengah_uas-1]
median_value_uas = (median_uas1 + median_uas2) / 2
print(f"mean : {mean_value_uas}")
print(f"median : {median_value_uas}\n")


counter_uas = Counter(data_uas)
mode_data_uas = dict(counter_uas)
max_counter_uas = max(list(mode_data_uas.values()))

mode_value_uas = []
for k, v in mode_data_uas.items() :
    print(f"{k} muncul {v} kali")
    if v == max_counter_uas :
        mode_value_uas.append(k)
        
print(f"{mode_value_uas}\n")
```

```python
plt.hist(data_uas, bins=30, edgecolor='black', alpha=0.5, color='blue')
plt.axvline(x=mean_value_uas, color='red', linestyle='dashed', linewidth=1, label=f'Mean = {mean_value_uas:.2f}')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Mean')

plt.legend()
plt.show()
```

```python
plt.hist(data_uas, bins=30, edgecolor='black', alpha=0.5, color='green')
plt.axvline(x=median_value_uas, color='red', linestyle='dashed', linewidth=1, label=f'Median = {median_value_uas:.2f}')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Median')

plt.legend()
plt.show()
```

```python
plt.hist(data_uas, bins=30, edgecolor='black', alpha=0.5, color='green')

#plt.axvline(x=mode_value[0], color='red', linestyle='dashed', linewidth=1, label=f'Modus 1 = {mode_value[0]:.2f}')

for i in mode_value_uas :
    plt.axvline(x=i, color='red', linestyle='dashed', linewidth=1, label=f'modus = {i:.2f}')


plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Modes')

plt.legend()
plt.show()
```
