```python
import matplotlib.pyplot as plt
import math as m
from collections import Counter

def calculate_standard_deviation(data):
    variance = calculate_variance(data)
    standard_deviation = variance**0.5
    
    return standard_deviation


def calculate_mean(data):
    return sum(data) / len(data)

def calculate_variance(data):
    mean = calculate_mean(data)
    squared_diff = []
    for x in data:
        squared_diff_per_x = (x - mean)**2
        squared_diff.append( squared_diff_per_x )  
    
    variance = sum(squared_diff) / len(data)   
    
    return variance

def calculate_range(data):
    nilai_tertinggi = max(data)
    nilai_terendah = min(data)
    
    data_range = nilai_tertinggi - nilai_terendah
    
    return data_range


data = [19, 10, 10, 12, 13, 15, 14, 13, 9, 17]
data_range = calculate_range(data)
data_variance = calculate_variance(data)
data_standard_deviation = calculate_standard_deviation(data)
mean = calculate_mean(data)
print(f"mean = {mean}")
print(f"range = {data_range}")
print(f"variance = {data_variance}")
print(f"standard deviation = {data_standard_deviation}")
```

```python
plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='green')

y = 1.25
x1 = 1/20*1
x2 = 19/20*1

plt.axhline(y=y, xmin=x1, xmax=x2, color='red', linestyle='dashed', linewidth=1, label = f'Range = {data_range}')
plt.axvline(x=min(data), color='blue', linestyle='dashed', linewidth=1, label = f' Min data = {min(data):.2f}')
plt.axvline(x=max(data), color='blue', linestyle='dashed', linewidth=1, label = f' Max data = {max(data):.2f}')

plt.xlabel('value')
plt.ylabel('frequency')
plt.title('range')

plt.legend()
plt.show()
```

```python
plt.hist(data, bins=30, edgecolor='black', alpha=0.5, color='green')

plt.axvline(x=mean, color='red', linestyle='dashed', linewidth=1, label = f'Mean = {mean:.2f}')
plt.axvline(x=mean-data_standard_deviation, color='blue', linestyle='dashed', linewidth=1, label = f'{mean-data_standard_deviation:.2f}')
plt.axvline(x=mean+data_standard_deviation, color='blue', linestyle='dashed', linewidth=1, label = f'{mean+data_standard_deviation:.2f}')

plt.title(f"std.Deviation {data_standard_deviation:.2f}")
plt.legend()
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.show()
```
